```python
# Importing The required Libraries for this Assignment
import pandas as pd
import requests
import pandas_ta as pt
```

```python
# As We have the API link of the data, we use the request Library to get the data on the API link

content = requests.get('https://www.alphavantage.co/query?function=TIME_SERIES_INTRADAY&symbol=IBM&interval=5min&apikey=demo')
```

```python
# We have successfully established connection with the API link

content
```

```python
# Now we have to parse the data into JSon format and save it

data = content.json()
```

```python
data
```

We have the Data but in a Json format. We cannot use this directly And make a dataframe. 

We need to get our required data from this and save it in the form of lists.

```python
# First we get all the keys as list inside the data variable. This will be the timestamp.

timestamp = list(data['Time Series (5min)'].keys())
```

```python
# Now similarly, we need to get the data that is further inside each timestamp key. For that, we will create empty lists of column name.

Open = list()
Close = list()
High = list()
Low = list()
Volume = list()
```

```python
# Now, we will interatively append each data inside the lists.

for i in timestamp:
    Open.append(data['Time Series (5min)'][i]['1. open'])
    Close.append(data['Time Series (5min)'][i]['4. close'])
    High.append(data['Time Series (5min)'][i]['2. high'])
    Low.append(data['Time Series (5min)'][i]['3. low'])
    Volume.append(data['Time Series (5min)'][i]['5. volume'])
    
```

```python
# Now that we have all our data in a readable format, we can make a dictionary so that it can be made into a Pandas dataframe.

all_data = {'Timestamp': timestamp, 
           'Open': Open,
           'Close': Close,
           'High': High,
           'Low': Low,
           "Volume": Volume}
```

```python
# Let's make the dataframe now.

df = pd.DataFrame(all_data)
```

```python
# As we can see, we have our dataframe with each column as we needed. 

df.head()
```

```python
# One thing we need to check before doing anything on this data is the datatype of each column.
# After checking, we can see that all the columns are of object datatype. We can't use these data unless they are in float or int datatype.

df.info()
```

```python
# Using astype() method, we can change the datatype of each column to float.

df['Open'] = df['Open'].astype('float')
df['Close'] = df['Close'].astype('float')
df['High'] = df['High'].astype('float')
df['Low'] = df['Low'].astype('float')
df['Volume'] = df['Volume'].astype('float')
```

```python
# Now, we can see that except the Timestamp column, the other columns are of float type datatype.

df.info()
```

```python
# Now, we will use the pandas_ta library which has a method named supertrend to find the supertrend.
# We will enter parameters such as high, low, and close as series from our df dataframe.

pt.supertrend(high=df['High'], low=df['Low'], close=df['Close'], period = 7, multiplier=3)
```

```python
# We only need the first column from this new dataframe. We will save it in our current dataframe.

df['SuperTrend'] = pt.supertrend(high=df['High'], low=df['Low'], close=df['Close'], period = 7, multiplier=3)['SUPERT_7_3.0']
```

```python
# Now we can see our updated Dataframe.

df.head(10)
```

```python
# Now, we need to set 2 new columns 'Buy' and 'Sell' in our dataframe.

df['Buy'] = 0
df['Sell'] = 0
```

```python
df.head()
```

```python
# Now, here comes our main work. Here we need to interatively check each row and the one before it. 
# We will start from Row 7, because there are NaN values till row 7. 


# So what happens here is that, if the previous closing value of the candle is less than or equal to the previous supertrend 
# value, at the same time the current closing value is greater than Supertrend value then, buy signal will be set to 1. That means to buy.


# Meanwhile, on the other hand, if the previous closing value is greater than or equal to the previous Supertrend value at the same
# time, current closing value is less than current supertrend value, then sell signal will be set to 1, which means to sell. 


n = 7
for i in range(n, len(df)):
    if df['Close'][i-1] <= df['SuperTrend'][i-1] and df['Close'][i] > df['SuperTrend'][i]:
        df['Buy'][i] = 1
    if df['Close'][i-1] >= df['SuperTrend'][i-1] and df['Close'][i] < df['SuperTrend'][i]:
        df['Sell'][i] = 1
```

```python
# here we can see on 92nd row, there is a buy signal. 

df.tail(20)
```
